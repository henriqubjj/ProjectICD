```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('Twitch_global_data.csv')
data
```

```python
ano_2020 = data.loc[data["year"] == 2020]
ano_2020
```

```python
sns.scatterplot(data=ano_2020, x="Hours_watched", y="Avg_viewers")
```

```python
data_group = ano_2020[["Hours_watched", "Avg_viewers"]]
data_group
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data_group)
```

```python
kmeans.labels_
```

```python
sns.scatterplot(data=data_group, x="Hours_watched", y="Avg_viewers", hue=kmeans.labels_)
```
